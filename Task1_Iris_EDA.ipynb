# Task 1: Exploring and Visualizing a Simple Dataset

## Objective
Learn how to load, inspect, and visualize a dataset to understand data trends and distributions.

## Dataset
**Iris Dataset** – A classic multivariate dataset containing 150 samples of iris flowers across 3 species (Setosa, Versicolor, Virginica), each described by 4 features: sepal length, sepal width, petal length, and petal width.

## Models Applied
No predictive model is trained in this task. The focus is purely on **Exploratory Data Analysis (EDA)** and visualization.

## Key Results and Findings
- The dataset is clean with no missing values.
- Petal length and petal width are the most discriminating features between species.
- Iris Setosa is linearly separable from the other two species.
- Versicolor and Virginica overlap slightly in feature space.

---
## 1. Import Libraries

```python
# Standard data manipulation and visualization libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# For loading the Iris dataset
from sklearn.datasets import load_iris

# Display settings
%matplotlib inline
sns.set_theme(style='whitegrid', palette='Set2')
print('Libraries imported successfully!')
```

---
## 2. Load the Dataset

```python
# Load the Iris dataset using seaborn (CSV-compatible format)
# Alternatively, it can be downloaded from UCI or loaded via sklearn
df = sns.load_dataset('iris')

print('Dataset loaded successfully!')
print(f'Shape: {df.shape}')
```

---
## 3. Data Inspection

```python
# View the first few rows of the dataset
print('--- First 5 Rows ---')
df.head()
```

```python
# Shape and column names
print(f'Shape (rows, columns): {df.shape}')
print(f'\nColumn Names: {list(df.columns)}')
```

```python
# Dataset info: data types, non-null counts
print('--- Dataset Info ---')
df.info()
```

---
## 4. Descriptive Statistics

```python
# Summary statistics for all numeric columns
print('--- Summary Statistics ---')
df.describe()
```

```python
# Check for missing values
print('--- Missing Values ---')
print(df.isnull().sum())
print(f'\nTotal missing values: {df.isnull().sum().sum()}')
```

```python
# Class distribution
print('--- Species Distribution ---')
print(df['species'].value_counts())
```

---
## 5. Data Visualization

### 5.1 Scatter Plot – Relationships Between Features

```python
# Scatter plot: Sepal Length vs Sepal Width colored by species
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Sepal features
sns.scatterplot(
    data=df, x='sepal_length', y='sepal_width',
    hue='species', ax=axes[0], s=80, edgecolor='black'
)
axes[0].set_title('Sepal Length vs Sepal Width', fontsize=13)
axes[0].set_xlabel('Sepal Length (cm)')
axes[0].set_ylabel('Sepal Width (cm)')

# Plot 2: Petal features
sns.scatterplot(
    data=df, x='petal_length', y='petal_width',
    hue='species', ax=axes[1], s=80, edgecolor='black'
)
axes[1].set_title('Petal Length vs Petal Width', fontsize=13)
axes[1].set_xlabel('Petal Length (cm)')
axes[1].set_ylabel('Petal Width (cm)')

plt.suptitle('Scatter Plots: Feature Relationships by Species', fontsize=15, y=1.02)
plt.tight_layout()
plt.savefig('scatter_plots.png', bbox_inches='tight', dpi=150)
plt.show()
print('Insight: Petal features clearly separate species, especially Setosa.')
```

### 5.2 Pair Plot – All Feature Relationships

```python
# Pair plot showing relationships across all feature combinations
pair_plot = sns.pairplot(df, hue='species', diag_kind='kde', plot_kws={'alpha': 0.7})
pair_plot.fig.suptitle('Pair Plot – All Feature Combinations by Species', y=1.02, fontsize=14)
plt.savefig('pair_plot.png', bbox_inches='tight', dpi=150)
plt.show()
print('Insight: Petal length & width are the strongest discriminating features.')
```

### 5.3 Histograms – Value Distributions

```python
# Histograms for each feature split by species
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.flatten()

for i, feature in enumerate(features):
    for species in df['species'].unique():
        subset = df[df['species'] == species][feature]
        axes[i].hist(subset, bins=15, alpha=0.6, label=species, edgecolor='black')
    axes[i].set_title(f'Distribution of {feature.replace("_", " ").title()}', fontsize=12)
    axes[i].set_xlabel('Value (cm)')
    axes[i].set_ylabel('Frequency')
    axes[i].legend()

plt.suptitle('Histograms: Feature Value Distributions by Species', fontsize=14)
plt.tight_layout()
plt.savefig('histograms.png', bbox_inches='tight', dpi=150)
plt.show()
print('Insight: Setosa has distinct smaller petal distributions compared to other species.')
```

### 5.4 Box Plots – Outlier Detection

```python
# Box plots for each feature to identify outliers
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.flatten()

for i, feature in enumerate(features):
    sns.boxplot(
        data=df, x='species', y=feature,
        ax=axes[i], palette='Set2'
    )
    axes[i].set_title(f'Box Plot: {feature.replace("_", " ").title()}', fontsize=12)
    axes[i].set_xlabel('Species')
    axes[i].set_ylabel('Value (cm)')

plt.suptitle('Box Plots: Feature Distributions and Outliers by Species', fontsize=14)
plt.tight_layout()
plt.savefig('box_plots.png', bbox_inches='tight', dpi=150)
plt.show()
print('Insight: A few outliers exist in sepal_width for Setosa, but overall the data is clean.')
```

### 5.5 Correlation Heatmap

```python
# Heatmap to visualize correlations between numeric features
plt.figure(figsize=(8, 6))
corr_matrix = df.drop(columns='species').corr()

sns.heatmap(
    corr_matrix, annot=True, fmt='.2f',
    cmap='coolwarm', linewidths=0.5, square=True
)
plt.title('Correlation Heatmap of Iris Features', fontsize=14)
plt.tight_layout()
plt.savefig('heatmap.png', bbox_inches='tight', dpi=150)
plt.show()
print('Insight: Petal length and petal width are highly positively correlated (r=0.96).')
```
